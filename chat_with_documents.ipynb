```python
#choose model
from transformers import AutoModelForCausalLM, BitsAndBytesConfig
model_id="Qwen/Qwen2.5-3B-Instruct"
```

```python
#load tokenizer
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained(model_id,trust_remote_code = True)
print("Tokenizer loaded successfully.")
```

```python
#load model with quantization
quantization_config = BitsAndBytesConfig(load_in_4bit = True,
                                         bnb_4bit_compute_dtype = torch.float16,  # or torch.bfloat16 if available
                                         bnb_4bit_quant_type = "nf4",  # normal float 4 quantization
                                         bnb_4bit_use_double_quant = True  # use nested quantization for more efficient memory usage
                                         )

model=AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config = quantization_config,
                                             device_map = "auto",
                                             trust_remote_code = True
)
```

```python
import os
from pypdf import PdfReader

pdf_path = "google_earning_transcript.pdf"

if os.path.exists(pdf_path):
    print(f"üìÑ Found PDF: {pdf_path}")
    reader = PdfReader(pdf_path)

    all_text = ""
    for i, page in enumerate(reader.pages, start=1):
        page_text = page.extract_text()
        if page_text:
            all_text += page_text + "\n"
        else:
            print(f"‚ö†Ô∏è No extractable text on page {i}")

    print("‚úÖ PDF text extracted.")
else:
    print(f"‚ùå PDF not found at {pdf_path}")
    all_text = ""
```

```python
print(all_text[:1000])
```

```python
MAX_CONTEXT_CHAR = 6000

def answer_question_from_pdf(document_text, llm_pipeline, question):


    if len(document_text) > MAX_CONTEXT_CHAR:
        print(f"‚ö†Ô∏è Document text ({len(document_text)} chars) exceeds limit ({MAX_CONTEXT_CHAR} chars). Truncating...")
        context = document_text[:MAX_CONTEXT_CHAR] + "..."
    else:
        context = document_text

    # Build prompt
    prompt_template = f"""<|system|>
You are an AI assistant. Answer the following question based *only* on the provided document text.
If the answer is not found in the document, say "The document does not contain information on this topic."
Do not use any prior knowledge.

Document Text:
---
{context}
---
<|end|>
<|user|>
Question: {question}<|end|>
<|assistant|>
Answer:"""

    print(f"\n--- Generating Answer for: '{question}' ---")

    # Run inference
    outputs = llm_pipeline(
        prompt_template,
        max_new_tokens=500,
        do_sample=True,
        temperature=0.2,
        top_p=0.9,
    )

    # Extract only the generated answer
    generated_text = outputs[0]['generated_text']
    answer_start_index = generated_text.find("Answer:") + len("Answer:")
    raw_answer = generated_text[answer_start_index:].strip()

    # Clean up unwanted tokens
    if "<|end|>" in raw_answer:
        raw_answer = raw_answer.split("<|end|>")[0].strip()

    print("--- Generation Complete ---")
    return raw_answer
```

```python
test_question="What is this document about?"
answer=answer_question_from_pdf(all_text,pipe,test_question)
print("\nTest Question:")
print_markdown(f"**Q:** {test_question}")
print("\nGenerated Answer:")
print_markdown(f"**A:** {answer}")
```
